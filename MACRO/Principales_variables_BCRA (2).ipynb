```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# IDs de las variables necesarias
ids = {
    "Circulacion": 17,
    "Cuenta_corriente": 22,
    "Caja_ahorro": 23
}

# Descargar los datos y armar DataFrames
dfs = []
for nombre, id_var in ids.items():
    url = f"https://api.bcra.gob.ar/estadisticas/v3.0/Monetarias/{id_var}"
    response = requests.get(url, verify=False)
    data = response.json()["results"]
    df = pd.DataFrame(data)
    df["fecha"] = pd.to_datetime(df["fecha"])
    df = df.sort_values("fecha")
    df["valor"] = pd.to_numeric(df["valor"], errors="coerce")
    df = df[["fecha", "valor"]].rename(columns={"valor": nombre})
    dfs.append(df)

# Unir por fecha
df_m2 = dfs[0]
for df in dfs[1:]:
    df_m2 = pd.merge(df_m2, df, on="fecha", how="outer")

# Calcular M2
df_m2["M2"] = df_m2[["Circulacion", "Cuenta_corriente", "Caja_ahorro"]].sum(axis=1)

# Graficar
plt.figure(figsize=(12,6))
plt.plot(df_m2["fecha"], df_m2["M2"])
plt.title("Evolución del M2 (Argentina)")
plt.xlabel("Fecha")
plt.ylabel("Millones de Pesos")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Desactivar warnings SSL (opcional)
import urllib3
urllib3.disable_warnings()

# Función para obtener series
def obtener_serie(id_variable):
    url = f"https://api.bcra.gob.ar/estadisticas/v3.0/Monetarias/{id_variable}"
    response = requests.get(url, verify=False)
    data = pd.DataFrame(response.json()["results"])
    data["fecha"] = pd.to_datetime(data["fecha"])
    data["valor"] = pd.to_numeric(data["valor"], errors='coerce')
    return data[["fecha", "valor"]]

# Descargar series
circulante = obtener_serie(17)
cuentas_corrientes = obtener_serie(22)
caja_ahorro = obtener_serie(23)
base_monetaria = obtener_serie(15)

# Renombrar columnas
circulante = circulante.rename(columns={"valor": "Circulante"})
cuentas_corrientes = cuentas_corrientes.rename(columns={"valor": "Ctas Corrientes"})
caja_ahorro = caja_ahorro.rename(columns={"valor": "Cajas Ahorro"})
base_monetaria = base_monetaria.rename(columns={"valor": "Base Monetaria"})

# Merge de todo
df = circulante.merge(cuentas_corrientes, on="fecha", how="inner")
df = df.merge(caja_ahorro, on="fecha", how="inner")
df = df.merge(base_monetaria, on="fecha", how="inner")

# Calcular depósitos a la vista y M2 transaccional
df["Depósitos Vista"] = df["Ctas Corrientes"] + df["Cajas Ahorro"]
df["M2 Privado"] = df["Circulante"] + df["Depósitos Vista"]

# Redondear a billones
df[["Circulante", "Depósitos Vista", "Base Monetaria", "M2 Privado"]] /= 1e6  # Millones a billones

# Filtrar últimos 16 meses
df = df.sort_values("fecha").set_index("fecha").resample("M").last().dropna().tail(16)

# Gráfico
plt.figure(figsize=(12,6))

# Barras apiladas
plt.bar(df.index, df["Circulante"], label="Circulante en poder del público", color="#009FE3")
plt.bar(df.index, df["Depósitos Vista"], bottom=df["Circulante"], label="Depósitos privados a la vista (excl. Rem)", color="#A4D4F4")

# Línea base monetaria
plt.plot(df.index, df["Base Monetaria"], label="Base Monetaria", color="yellow", linestyle="--", linewidth=2)

# Línea M2 transaccional
plt.plot(df.index, df["M2 Privado"], label="M2 transaccional privado", color="#004C70", linewidth=2)

# (Opcional) Objetivo BCRA (simulado manualmente)
objetivo_bcra = pd.Series([52.3, 57.0, 65.8], index=pd.to_datetime(["2025-06-30", "2025-07-31", "2025-10-31"]))
plt.scatter(objetivo_bcra.index, objetivo_bcra.values, label="Objetivo BCRA", marker='D', color="teal", zorder=5)

# Estética
plt.title("Agregados Monetarios - M2 Transaccional")
plt.ylabel("Billones de pesos")
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
