```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# IDs de las variables que quieres descargar
ids = [6, 45]
nombres = {6: "TMA", 45: "TAMAR"}

# Diccionario para guardar los DataFrames de cada variable
dfs = {}

# Fecha de inicio (octubre 2024)
fecha_inicio = pd.to_datetime("2024-10-01")

for id_var in ids:
    url = f"https://api.bcra.gob.ar/estadisticas/v3.0/Monetarias/{id_var}"
    response = requests.get(url, verify=False)
    data = response.json()["results"]
    df = pd.DataFrame(data)
    df["fecha"] = pd.to_datetime(df["fecha"])
    df = df[df["fecha"] >= fecha_inicio]  # Filtrar desde octubre 2024
    df = df.sort_values("fecha")
    dfs[id_var] = df

# Graficar ambas variables
plt.figure(figsize=(12,6))
for id_var, df in dfs.items():
    plt.plot(df["fecha"], df["valor"], label=nombres[id_var])

plt.title("TPM vs TAMAR (Tasas en pesos)")
plt.xlabel("Fecha")
plt.ylabel("Tasa %")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# IDs de las variables que quieres descargar
ids = [6, 45]
nombres = {6: "TPM", 45: "TAMAR"}

# Diccionario para guardar los DataFrames de cada variable
dfs = {}

# Fecha de inicio (octubre 2024)
fecha_inicio = pd.to_datetime("2024-10-01")

for id_var in ids:
    url = f"https://api.bcra.gob.ar/estadisticas/v3.0/Monetarias/{id_var}"
    response = requests.get(url, verify=False)
    data = response.json()["results"]
    df = pd.DataFrame(data)
    df["fecha"] = pd.to_datetime(df["fecha"])
    df = df[df["fecha"] >= fecha_inicio]  # Filtrar desde octubre 2024
    df = df.sort_values("fecha")
    # Convertir tasa anual a mensual
    df["tasa_mensual"] = ((1 + df["valor"]/100)**(1/12) - 1) * 100
    dfs[id_var] = df

# Graficar ambas variables (tasa mensual)
plt.figure(figsize=(12,6))
for id_var, df in dfs.items():
    plt.plot(df["fecha"], df["tasa_mensual"], label=nombres[id_var])

plt.title("TPM vs TAMAR (Tasas mensuales equivalentes en pesos)")
plt.xlabel("Fecha")
plt.ylabel("Tasa mensual (%)")
plt.legend()
plt.grid(True)
plt.show()
```
